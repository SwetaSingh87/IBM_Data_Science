```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
launch_sites = {
    "Cape Canaveral": [28.56230197, -80.57735648],
    "Vandenberg": [34.63283416, -120.6107455],
    "Kennedy Space Center": [28.57325457, -80.64689529],
}

# Sample point of interest coordinates
points_of_interest = {
    "City": [29.7604, -95.3698],  # Houston, TX
    "Railway": [29.7604, -95.3698],  # Example railway coordinates
    "Highway": [29.7604, -95.3698],  # Example highway coordinates
}

# Create a map object
mymap = folium.Map(location=[30, -95], zoom_start=4)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(mymap)

# Iterate through each launch site
for launch_site, launch_coord in launch_sites.items():
    # Initialize minimum distance and closest point of interest
    min_distance = float('inf')
    closest_point = None
    
    # Iterate through each point of interest
    for poi_name, poi_coord in points_of_interest.items():
        # Calculate distance between launch site and point of interest
        distance = ((launch_coord[0] - poi_coord[0])**2 + (launch_coord[1] - poi_coord[1])**2)**0.5
        
        # Update closest point of interest if distance is smaller
        if distance < min_distance:
            min_distance = distance
            closest_point = poi_name
    
    # Add Marker to MarkerCluster with popup showing launch site and distance to closest point of interest
    popup_content = f"Launch Site: {launch_site}<br>Distance to {closest_point}: {min_distance:.2f} units"
    folium.Marker(location=launch_coord, popup=popup_content).add_to(marker_cluster)
    
    # Draw a line between the marker and the launch site
    folium.PolyLine(locations=[launch_coord, points_of_interest[closest_point]], color='blue').add_to(mymap)

# Display the map
mymap
```
